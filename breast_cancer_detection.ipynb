# ***KNN***

```python
def euclid_dist(data_point_one,data_point_two):
  return np.sqrt(np.sum((data_point_one-data_point_two)**2))
def accuracy(y_actual_value, y_predicted_value):
        accuracy = np.sum(y_actual_value == y_predicted_value) / len(y_actual_value)
        return accuracy
class KNN:
  def __init__(self,no_of_neighbors=1):
    self.no_of_neighbors = no_of_neighbors

  def fit(self,X,y):
    self.X_train_data = X
    self.y_train_data = y

  def predict(self,X_test_data):
    y_pred_data = [self.__predict(x_test_data) for x_test_data in X_test_data]
    return np.array(y_pred_data)
  def __predict(self,x):
    #compute distance between given testing data point and all other training data
    distances= [euclid_dist(x,x_train) for x_train in self.X_train_data]
    # sort distances from all the points and take nearest k
    k_ind= np.argsort(distances)[:self.no_of_neighbors]
    #extract labels of k nearest neighbor
    k_neigh_labels= [self.y_train_data[i] for i in k_ind]
    # return most common class label
    most_common_target_name = Counter(k_neigh_labels).most_common(1)
    #print(most_common_target_name)
    return most_common_target_name[0][0]
```

# ***Logistic Regression***

```python
class LogisticRegression:
  def __init__(self,learning_rate=0.001, iterations=1000):
      self.lr=learning_rate
      self.iters=iterations
      self.W=None
      self.B=None

  def gradient(self,m,X,y,A):
    dW=(1/m)*np.dot(X.T,A-y)
    dB=(1/m)*np.sum(A - y)
    self.W-=self.lr*dW.T
    self.B-=self.lr*dB
    return self.W,self.B

  def fit(self,X,y):
    m,features=X.shape
    self.W=np.zeros(features)
    self.B=0
    for _ in range(self.iters):
      Z=np.dot(X,self.W)+self.B
      A=self.sigmoid(Z)
      self.W,self.B=self.gradient(m,X,y,A)

  def sigmoid(self,x):
    sig=1/(1+exp(-x))
    return sig

  def predict(self,X):
      Z=np.dot(X,self.W)+self.B
      A=self.sigmoid(Z)
      predicted_vals = [1 if i > 0.5 else 0 for i in A]
      return np.array(predicted_vals)
```

#**Naiyes Bayes**


```python
class Gaussian_Naive_Bayes_Classifier:
    def fit(self, Features, Target):
        num_data, num_features = len(Features), len(Features[0])
        self.classes = np.unique(Target)
        num_classes = len(self.classes)
        self.MLis = np.zeros((num_classes, num_features), dtype=np.float64)
        self.VLis = np.zeros((num_classes, num_features), dtype=np.float64)
        self.priors = np.zeros(num_classes, dtype=np.float64)
        for i in self.classes:
            dis = Features[Target == i]
            self.MLis[i, :] = dis.mean(axis=0)
            self.VLis[i, :] = dis.var(axis=0)
            self.priors[i] = dis.shape[0] / float(num_data)
    def predict(self, Features):
        pred = [self.class_prob(x) for x in Features]
        return np.array(pred)
    def class_prob(self, x):
        posteriors = []
        for i in self.classes:
            prior = np.log(self.priors[i])
            ans = self.density(i, x)
            for i in range(len(ans)):
              if ans[i] == 0:
                ans[i] = 1
            posterior = np.sum(np.log(ans))
            posterior = prior + posterior
            posteriors.append(posterior)
        return self.classes[np.argmax(posteriors)]
    def density(self, i, x):
        mean = self.MLis[i]
        var = self.VLis[i]
        numerator = np.exp(-((x - mean) ** 2) / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator
```

```python
def breastCancerTest():
    NB = Gaussian_Naive_Bayes_Classifier()
    NB.fit(X_train, y_train)
    naive_bayes_pred = NB.predict(X_test)
    def accuracy(y_true, y_pred):
        accuracy = np.sum(y_true == y_pred) / len(y_true)
        return accuracy
    print("Gaussian Naive Bayes accuracy: {:.2f}".format(accuracy(y_test, naive_bayes_pred)))
    return naive_bayes_pred
```
